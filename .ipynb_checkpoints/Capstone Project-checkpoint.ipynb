```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('kag_risk_factors_cervical_cancer.xls')
pd.set_option('display.max_columns', 37)
df.shape
```

```python
df.columns
```

```python
l = ['Age', 'Number of sexual partners', 'First sexual intercourse',
       'Num of pregnancies', 'Smokes', 'Smokes (years)', 'Smokes (packs/year)',
       'Hormonal Contraceptives', 'Hormonal Contraceptives (years)', 'IUD',
       'IUD (years)', 'STDs', 'STDs (number)', 'STDs:condylomatosis',
       'STDs:cervical condylomatosis', 'STDs:vaginal condylomatosis',
       'STDs:vulvo-perineal condylomatosis', 'STDs:syphilis',
       'STDs:pelvic inflammatory disease', 'STDs:genital herpes',
       'STDs:molluscum contagiosum', 'STDs:AIDS', 'STDs:HIV',
       'STDs:Hepatitis B', 'STDs:HPV', 'STDs: Number of diagnosis',
       'STDs:Time since first diagnosis', 'STDs:Time since last diagnosis',
       'Dx:Cancer', 'Dx:CIN', 'Dx:HPV', 'Dx', 'Hinselmann', 'Schiller',
       'Citology', 'Biopsy']
l1 = []
l2 = []
l3 =[]
l4 = []
l5 = []
for i in l:
    j = i.split(' ')
    k = '_'.join(j)
    l1.append(k)
for i in l1:
    m = i.split(':')
    k1 = '_'.join(m)
    l2.append(k1)
for i in l2:
    n = i.split('(')
    k2 = ''.join(n)
    l3.append(k2)
for i in l3:
    n1 = i.rstrip(')')
    l4.append(n1)
for i in l4:
    n2 = i.split('-')
    k3 = '_'.join(n2)
    l5.append(k3)

df.columns = l5
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().mean()*100
```

```python
df.replace('?', np.NaN, inplace=True)
df.info()
```

```python
df.isnull().mean()*100
```

```python
df=df.drop(['STDs_Time_since_first_diagnosis','STDs_Time_since_last_diagnosis'],axis=1)
```

```python
df = df.drop(df.index[df.Smokes.isnull()] | df.index[df.First_sexual_intercourse.isnull()])
```

```python
x_features = list(df.columns)
x_features.remove('Biopsy')

x_features_cat = [
    'Smokes','Hormonal_Contraceptives','IUD','STDs','STDs_condylomatosis','STDs_cervical_condylomatosis',
    'STDs_vaginal_condylomatosis','STDs_vulvo_perineal_condylomatosis','STDs_syphilis','STDs_pelvic_inflammatory_disease',
    'STDs_genital_herpes','STDs_molluscum_contagiosum','STDs_AIDS','STDs_HIV','STDs_Hepatitis_B','STDs_HPV',
    'Dx_Cancer','Dx_CIN','Dx_HPV','Dx', 'Hinselmann', 'Schiller', 'Citology']

x_features_cat
```

```python
x_features_num = [i for i in x_features if i not in x_features_cat]
x_features_num
```

```python
df[['Schiller', 'Hinselmann', 'Citology', 'Biopsy']][df[['Schiller', 'Hinselmann', 'Citology', 'Biopsy']] == 1].count()
```

```python
df['Biopsy'].value_counts()
```

```python
### Missing Value Imputation of IUD

df_iud=df.copy()

df_iud.info()
```

```python
x_features_cat.remove('IUD')

x_features_cat
```

```python
for i in x_features_cat:
    df_iud[i] = df_iud[i].fillna(df_iud[i].mode()[0])

for i in x_features_num:
    df_iud[i] = df_iud[i].fillna(df_iud[i].median())

df_iud = df_iud.astype('float')
df_iud[x_features_cat] = df_iud[x_features_cat].replace(0,'no')
df_iud[x_features_cat] = df_iud[x_features_cat].replace(1,'yes')
df_iud.info()
```

```python
df_iud.head()
```

```python
cat = df_iud[x_features_cat].apply(LabelEncoder().fit_transform)
num = df_iud.drop(x_features_cat, axis=1)
df_iud1 = pd.concat([num, cat], axis = 1)
df_iud1.head()
df_iud1.shape
```

```python
df_iud=pd.get_dummies(df_iud)
df_iud.head()
df_iud.shape
```

```python
train_iud=df_iud[df_iud.IUD.isnull()==False]
train_iud.shape
```

```python
test_iud=df_iud[df_iud.IUD.isnull()]
test_iud.shape
```

```python
train_iud_x=train_iud.drop('IUD',axis=1)
train_iud_y=train_iud['IUD']
```

```python
test_iud_x=test_iud.drop('IUD',axis=1)
test_iud_y=test_iud['IUD']
```

```python
dt=DecisionTreeClassifier()
iud_model=dt.fit(train_iud_x,train_iud_y)
iud_model.score(train_iud_x,train_iud_y)
```

```python
test_iud['IUD']=iud_model.predict(test_iud_x)

df_impute=df.copy()

iud_complete=pd.concat([train_iud,test_iud],axis=0)
df_impute['IUD']=iud_complete['IUD'].sort_index()
df_impute.info()
```

```python
x_features_num.remove('IUD_years')
x_features_num

df_impute[df_impute['IUD_years']==0][['IUD','IUD_years']].sum()
df_impute[df_impute['IUD']==0][['IUD','IUD_years']].sum()
df_impute['IUD_years']=df_impute['IUD_years'].fillna(0)
df_impute.info()
```

```python
df_std_num=df_impute.drop(['Biopsy'],axis=1)

x_features_num.remove('STDs_number')
x_features_num
```

```python
for i in x_features_cat:
    df_std_num[i]=df_std_num[i].fillna(df_std_num[i].mode()[0])

for i in x_features_num:
    df_std_num[i]=df_std_num[i].fillna(df_std_num[i].median())

df_std_num=df_std_num.astype('float')
df_std_num[x_features_cat]=df_std_num[x_features_cat].replace(0,'no')
df_std_num[x_features_cat]=df_std_num[x_features_cat].replace(1,'yes')

df_std_num=pd.get_dummies(df_std_num)
df_std_num.head()
```

```python
train_std_num=df_std_num[df_std_num.STDs_number.isnull()==False]
test_std_num=df_std_num[df_std_num.STDs_number.isnull()]

train_std_num_x=train_std_num.drop('STDs_number',axis=1)
train_std_num_y=train_std_num['STDs_number']

test_std_num_x=test_std_num.drop('STDs_number',axis=1)
test_std_num_y=test_std_num['STDs_number']

dt=DecisionTreeRegressor()
std_model_num=dt.fit(train_std_num_x,train_std_num_y)
std_model_num.score(train_std_num_x,train_std_num_y)
```

```python
test_std_num['STDs_number']=std_model_num.predict(test_std_num_x)
std_model_num.predict(test_std_num_x)

std_num_complete=pd.concat([train_std_num,test_std_num],axis=0)
df_std_num['STDs_number']=std_num_complete['STDs_number'].sort_index()
```

```python
df_impute['STDs_number']=df_std_num['STDs_number']
df_impute.info()
```
